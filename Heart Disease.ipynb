```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings

df = pd.read_csv("/kaggle/input/heart-disease-statlog/Heart_disease_statlog.csv")

warnings.simplefilter(action='ignore', category=FutureWarning)

# Set seaborn style and preferred palette
sns.set(style="whitegrid")
palette_seq = "Blues_r"
palette_qual = "bright"

# Plot 1: Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['age'], kde=True, bins=20, color=sns.color_palette(palette_seq)[2])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 2: Bar chart of Chest Pain Type (cp) by frequency
plt.figure(figsize=(8, 6))
cp_counts = df['cp'].value_counts().sort_values(ascending=False)
sns.barplot(x=cp_counts.index, y=cp_counts.values, palette=palette_seq)
plt.title("Chest Pain Type Frequency")
plt.xlabel("Chest Pain Type (cp)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Plot 3: Target variable count (Heart Disease presence)
plt.figure(figsize=(8, 6))
target_counts = df['target'].value_counts().sort_values(ascending=False)
sns.barplot(x=target_counts.index, y=target_counts.values, palette=palette_seq)
plt.title("Heart Disease Presence")
plt.xlabel("Target (0 = No Disease, 1 = Disease)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Plot 4: Heart disease by sex
plt.figure(figsize=(8, 6))
sns.countplot(x="sex", hue="target", data=df, palette=palette_qual)
plt.title("Heart Disease Presence by Sex")
plt.xlabel("Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.legend(title="Heart Disease")
plt.tight_layout()
plt.show()

# Plot 5: Maximum Heart Rate Achieved (thalach) distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['thalach'], kde=True, bins=20, color=sns.color_palette(palette_seq)[3])
plt.title("Maximum Heart Rate (thalach) Distribution")
plt.xlabel("thalach")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 6: Resting Blood Pressure (trestbps) distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['trestbps'], kde=True, bins=20, color=sns.color_palette(palette_seq)[4])
plt.title("Resting Blood Pressure (trestbps) Distribution")
plt.xlabel("trestbps")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 7: Serum Cholesterol (chol) distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['chol'], kde=True, bins=20, color=sns.color_palette(palette_seq)[5])
plt.title("Serum Cholesterol (chol) Distribution")
plt.xlabel("chol")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Re-plot Plot 8 with corrected color index
plt.figure(figsize=(8, 6))
sns.histplot(df['oldpeak'], kde=True, bins=20, color=sns.color_palette(palette_seq)[-1])
plt.title("ST Depression Induced by Exercise (oldpeak)")
plt.xlabel("oldpeak")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 9: Bar chart of resting electrocardiographic results (restecg)
plt.figure(figsize=(8, 6))
restecg_counts = df['restecg'].value_counts().sort_values(ascending=False)
sns.barplot(x=restecg_counts.index, y=restecg_counts.values, palette=palette_seq)
plt.title("Resting ECG Results (restecg)")
plt.xlabel("restecg (0 = Normal, 1 = ST-T abnormality, 2 = LV hypertrophy)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

sns.pairplot(df, hue = 'target',palette = 'mako' )

sns.pairplot(df, hue = 'target', vars = ['age', 'trestbps', 'chol'], palette ='bright' )
```
